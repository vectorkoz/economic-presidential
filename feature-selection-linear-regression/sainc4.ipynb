```python
# Surpress warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import numpy as np
import pandas as pd
```

```python
sainc4=pd.read_csv('data/income/SAINC4__ALL_AREAS_1929_2020.csv')
sainc4.drop(columns=['IndustryClassification','Region'],inplace=True)
sainc4=sainc4[~(sainc4.isnull().any(axis=1))]
```

```python
sainc4.drop(columns=[str(x) for x in np.arange(1929,1976)],inplace=True)
```

```python
sainc4.info()
```

```python
sainc4.rename(columns={'GeoFIPS':'fips'},inplace=True)
sainc4=sainc4.astype({'fips': 'str'})
sainc4.fips = sainc4.fips.apply(lambda x: x.strip()[1:-1])
sainc4=sainc4.astype({'fips': 'int'})
sainc4.fips = sainc4.fips.apply(lambda x: int(x/1000))
sainc4.fips.unique()
```

```python
sainc4=sainc4[sainc4.fips<90]
```

```python
sainc4.rename(columns={'GeoName':'state'},inplace=True)
```

```python
sainc4[['LineCode','Description','Unit']].drop_duplicates()
```

```python
def str_transform(feature,unit):
    feature=feature.strip()
    if feature[-1]=='/':
        feature=feature[:-3]
    return feature+', '+unit.strip()
```

```python
sainc4.Description=pd.Series([str_transform(x,y) for x,y in zip(sainc4.Description,sainc4.Unit)])
```

```python
sainc4.rename(columns={'LineCode':'code','Description':'features'},inplace=True)
sainc4=sainc4.astype({'code': 'int'})
sainc4.drop(columns=['Unit'],inplace=True)
```

```python
sainc4=sainc4[sainc4.features!='Population (persons), Number of persons']
```

```python
sainc4.info()
```

```python
def attempt_conversion(x):
    try:
        float(x)
    except:
        return True
    else:
        return False
```

```python
sainc4[sainc4['1983'].apply(attempt_conversion)]
```

```python
sainc4=sainc4[sainc4.features!="Farm proprietors' income, Millions of dollars"]
```

```python
sainc4=sainc4.astype({'1983':'float'})
```

```python
sainc4.info()
```
