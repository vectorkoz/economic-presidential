```python
# Surpress warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import numpy as np
import pandas as pd
```

```python
sainc30=pd.read_csv('data/income/SAINC30__ALL_AREAS_1958_2020.csv')
sainc30.drop(columns=['IndustryClassification','Region'],inplace=True)
sainc30=sainc30[~(sainc30.isnull().any(axis=1))]
```

```python
sainc30.info()
```

```python
sainc30.drop(columns=[str(x) for x in np.arange(1958,1976)],inplace=True)
```

```python
sainc30.info()
```

```python
sainc30.GeoFIPS.unique()
```

```python
sainc30.rename(columns={'GeoFIPS':'fips'},inplace=True)
sainc30=sainc30.astype({'fips': 'str'})
sainc30.fips = sainc30.fips.apply(lambda x: x.strip()[1:-1])
sainc30=sainc30.astype({'fips': 'int'})
sainc30.fips = sainc30.fips.apply(lambda x: int(x/1000))
sainc30.fips.unique()
```

```python
sainc30=sainc30[sainc30.fips<90]
```

```python
sainc30.rename(columns={'GeoName':'state'},inplace=True)
```

```python
sainc30[['LineCode','Description','Unit']].drop_duplicates()
```

```python
def str_transform(feature,unit):
    feature=feature.strip()
    if feature[-1]=='/':
        feature=feature[:-3]
    return feature+', '+unit.strip()
```

```python
sainc30.Description=pd.Series([str_transform(x,y) for x,y in zip(sainc30.Description,sainc30.Unit)])
```

```python
sainc30[['LineCode','Description','Unit']].drop_duplicates()
```

```python
sainc30.rename(columns={'LineCode':'code','Description':'features'},inplace=True)
sainc30=sainc30.astype({'code': 'int'})
sainc30.drop(columns=['Unit'],inplace=True)
```
