```python
# Surpress warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import numpy as np
import pandas as pd
```

```python
sainc50=pd.read_csv('data/income/SAINC50__ALL_AREAS_1948_2020.csv')
sainc50.drop(columns=['IndustryClassification','Region'],inplace=True)
sainc50=sainc50[~(sainc50.isnull().any(axis=1))]
```

```python
sainc50.head()
```

```python
sainc50.info()
```

```python
sainc50.drop(columns=[str(x) for x in np.arange(1948,1976)],inplace=True)
```

```python
sainc50.info()
```

```python
sainc50.GeoFIPS.unique()
```

```python
sainc50.rename(columns={'GeoFIPS':'fips'},inplace=True)
sainc50=sainc50.astype({'fips': 'str'})
sainc50.fips = sainc50.fips.apply(lambda x: x.strip()[1:-1])
sainc50=sainc50.astype({'fips': 'int'})
sainc50.fips = sainc50.fips.apply(lambda x: int(x/1000))
```

```python
sainc50.fips.unique()
```

```python
sainc50[['fips','GeoName']].drop_duplicates()
```

```python
sainc50=sainc50[sainc50.fips<90]
```

```python
sainc50.rename(columns={'GeoName':'state'},inplace=True)
```

```python
sainc50.info()
```

```python
sainc50[['LineCode','Description','Unit']].drop_duplicates()
```

```python
def str_transform(feature,unit):
    feature=feature.strip()
    if feature[-1]=='/':
        feature=feature[:-3]
    return feature+', '+unit.strip()
```

```python
sainc50.Description=pd.Series([str_transform(x,y) for x,y in zip(sainc50.Description,sainc50.Unit)])
```

```python
sainc50[['LineCode','Description','Unit']].drop_duplicates()
```

```python
sainc50.drop(columns=['Unit'],inplace=True)
```

```python
sainc50.rename(columns={'LineCode':'code','Description':'features'},inplace=True)
sainc50=sainc50.astype({'code': 'int'})
```

```python
sainc50=sainc50[sainc50.features!='Population (persons), Number of persons']
```

```python
sainc50.info()
```

```python
sainc50.describe()
```
