```python
# Surpress warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import numpy as np
import pandas as pd
```

```python
sainc35=pd.read_csv('data/income/SAINC35__ALL_AREAS_1929_2020.csv')
sainc35.drop(columns=['IndustryClassification','Region'],inplace=True)
sainc35=sainc35[~(sainc35.isnull().any(axis=1))]
```

```python
sainc35.drop(columns=[str(x) for x in np.arange(1929,1976)],inplace=True)
```

```python
sainc35.info()
```

```python
sainc35.GeoFIPS.unique()
```

```python
sainc35.rename(columns={'GeoFIPS':'fips'},inplace=True)
sainc35=sainc35.astype({'fips': 'str'})
sainc35.fips = sainc35.fips.apply(lambda x: x.strip()[1:-1])
sainc35=sainc35.astype({'fips': 'int'})
sainc35.fips = sainc35.fips.apply(lambda x: int(x/1000))
sainc35.fips.unique()
```

```python
sainc35=sainc35[sainc35.fips<90]
```

```python
sainc35.rename(columns={'GeoName':'state'},inplace=True)
```

```python
sainc35[['LineCode','Description','Unit']].drop_duplicates()
```

```python
def str_transform(feature,unit):
    feature=feature.strip()
    if feature[-1]=='/':
        feature=feature[:-3]
    return feature+', '+unit.strip()
```

```python
sainc35.Description=pd.Series([str_transform(x,y) for x,y in zip(sainc35.Description,sainc35.Unit)])
```

```python
sainc35.rename(columns={'LineCode':'code','Description':'features'},inplace=True)
sainc35=sainc35.astype({'code': 'int'})
sainc35.drop(columns=['Unit'],inplace=True)
```

```python
sainc35.info()
```

```python
unavailable_features=sainc35[(sainc35=='(NA)').any(axis=1)].features.unique().tolist()
unavailable_features
```

```python
sainc35=sainc35[~(sainc35.features.isin(unavailable_features))]
```

```python
sainc35.info()
```

```python
def attempt_conversion(x):
    try:
        float(x)
    except:
        return True
    else:
        return False
```

```python
sainc35[sainc35['1983'].apply(attempt_conversion)].size
```

```python
for year in np.arange(1976,2003):
    if (sainc35[sainc35['1983'].apply(attempt_conversion)].size!=0):
        print(year)
```

```python
conversion_dict={str(x):'float' for x in np.arange(1976,2003)}
```

```python
sainc35=sainc35.astype(conversion_dict)
```
